## Section 3: Finding Correlations

Lets start by breaking our data into two.  

```python
# splitting the dataframe into 2 parts
# on basis of ‘DEATH_EVENT’ column values
df_died = data[data['DEATH_EVENT'] == 1 ]
df_lived = data[data['DEATH_EVENT'] == 0 ]
```

```python
# Plot your variable based on if they died or lived

sns.distplot(df_died['age'])
sns.distplot(df_lived['age'])
plt.title("Chances of survival vs Age")
plt.legend(('Died','Lived'))
plt.plot()
```

```python
# Feel free to add any additional graphs that help you answer the questions below.
sns.countplot(data=df_lived, x='diabetes', palette='pastel')
plt.title("Diabetes Prevelance Among Survivors")
```

```python
sns.countplot(data=df_died, x='diabetes', palette='pastel')
plt.title("Diabetes Prevelance Among Deceased")
```

#### What things can you infer if we consider our data a sample of the population, based on each of your variables.  
<br><br><br><br><br>
#### Do you think either of your variables is a good indicator for predicting Heart Failure, why or why not?  
<br><br><br><br><br>

```python
#DIABETES DOES NOT PREDICT A HEART RELATED DEATH EVENT
import scipy.stats as stats

# Create a contingency table
contingency_table = pd.crosstab(data['DEATH_EVENT'], data['diabetes'])

# Perform a chi-square test
chi2_stat, p_value, dof, expected = stats.chi2_contingency(contingency_table)

# Print the results
print("Chi-square statistic:", chi2_stat)
print("P-value:", p_value)
print("Degrees of freedom:", dof)
print("Expected frequencies table:")
print(expected)

# Check the p-value to determine significance
alpha = 0.05  # Set your significance level
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant association between the variables.")
else:
    print("Fail to reject the null hypothesis: There is no significant association between the variables.")
```

```python
#THERE IS NOT A SIGNIFICANT DIFFERENCE BETWEEN THE AVERAGE AGES OF THOSE WHO SURVIVED AND THOSE WHO DIED

import scipy.stats as stats

# Extract the data from the DataFrame
variable1_data = df_lived['age']
variable2_data = df_died['age']

# Perform the independent t-test
t_stat, p_value = stats.ttest_ind(variable1_data, variable2_data)

# Print the results
print("T-statistic:", t_stat)
print("P-value:", p_value)

# Check the p-value to determine significance
alpha = 0.05  # Set your significance level
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the variables.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the variables.")
    
print(f"Mean age of those who lived: {round(df_lived['age'].mean(),2)}")
print(f"Mean age of those who died: {round(df_died['age'].mean(), 2)}")
```

```python
#BUT AGE DOES PREDICT A DEATH EVENT - OLDER PERSONS FROM SAMPLE ARE STATISTICALLY MORE LIKLEY TO HAVE DIED (obviously...)
import statsmodels.api as sm

# Add an intercept term (constant) for logistic regression
data['Intercept'] = 1

# Define the predictor variables (continuous predictor and intercept)
X = data[['age', 'Intercept']]

# Define the outcome variable (dichotomous outcome)
y = data['DEATH_EVENT']

# Fit a logistic regression model
logit_model = sm.Logit(y, X)

# Perform the logistic regression and get the summary
logit_result = logit_model.fit()
print(logit_result.summary())
```
